```python
import pandas as pd
results_df = pd.read_csv("Rahul RAG llm evaluationmetric results_df.csv")
```

```python
results_df
```

```python
# Ensure 'is_adherent' is numeric (if it's not already)
results_df['is_adherent'] = results_df['is_adherent'].astype(int)  # Assuming binary values

# Group by dataset_name and calculate averages for each metric
average_metrics_by_dataset = results_df.groupby('dataset_name')[[
    'context_relevance',
    'utilization_score',
    'completeness_score',
    'is_adherent'
]].mean().reset_index()

print(average_metrics_by_dataset)
```

```python
#df to csv
average_metrics_by_dataset.to_csv("RAG grp2 average_metrics_by_dataset.csv", index=False)
```
